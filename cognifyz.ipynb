# **TASK 1**

**Predict Restaurant Ratings**

**Objective:**-

Build a machine learning model to predict restaurant ratings.

Steps:-


1.   Data preprocessing
2.   Feature encoding
3.   Train-test split
4.   Model training (Regression)
5.   Evaluation

**Step 1: Import Libraries**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

**Step 2: Load Dataset**

```python
df = pd.read_csv(r"C:\Users\0aaso\Desktop\csv\Dataset _cognifyz.csv")
df.head()
```

**Step 3: Preprocessing**

```python
df.fillna(method='ffill', inplace=True)

# Encode categorical columns
le = LabelEncoder()
for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])
```

**Step 4: Train-Test Split**

```python
X = df.drop('Aggregate rating', axis=1)
y = df['Aggregate rating']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**Step 5: Model Training**

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

**Step 6: Evaluation**

```python
y_pred = model.predict(X_test)

print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

# TASK 2

**Restaurant Recommendation System**

**Objective**:-
Recommend restaurants based on user preferences.

**Steps**:-

1.   Data preprocessing
1.   Define recommendation criteria
3.   Content-based filtering

**Step 1: Import**

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

**Step 2: Feature Selection**

```python
features = df[['Cuisines', 'Price range']]

# Convert text to numeric
features = pd.get_dummies(features)
```

**Step 3: Similarity Matrix**

```python
similarity = cosine_similarity(features)
```

**Step 4: Recommendation Function**

```python
def recommend(index):
    scores = list(enumerate(similarity[index]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    for i in scores[1:6]:
        print(df.iloc[i[0]]['Restaurant Name'])
```

**Step 5: Test**

```python
recommend(0)
```

# TASK 3

**Cuisine Classification**

**Objective**:-
Classify restaurants based on cuisine.

**Steps**:-


1.   Preprocessing
1.   Feature encoding
3.   Classification model
4.   Evaluation

**Step 1: Import**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

**Step 2: Encoding**

```python
le = LabelEncoder()
df['Cuisines'] = le.fit_transform(df['Cuisines'])

for col in df.select_dtypes(include='object').columns:
    df[col] = le.fit_transform(df[col])
```

**Step 3: Train-Test Split**

```python
X = df.drop('Cuisines', axis=1)
y = df['Cuisines']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

**Step 4: Model Training**

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

Step 5: Evaluation

```python
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# TASK 4

**Location-Based Analysis**

**Objective**:-
Analyze restaurant distribution based on location.

**Steps**:-

1.   Map visualization
2.   City-wise grouping
3.   Rating & price analysis

**Step 1: Import**

```python
import folium
from folium.plugins import MarkerCluster
```

**Step 2: Load Data**

```python
df = pd.read_csv(r"C:\Users\0aaso\Desktop\csv\Dataset _cognifyz.csv")
df = df.dropna(subset=['Latitude', 'Longitude'])
```

**Step 3: Map**

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import webbrowser

# Load dataset (your path)
path = r"C:\Users\0aaso\Desktop\csv\Dataset _cognifyz.csv"
df = pd.read_csv(path)

# Drop missing coordinates
df = df.dropna(subset=['Latitude', 'Longitude'])

# Create map
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=5)

# Add cluster
cluster = MarkerCluster().add_to(m)

# Add markers
for _, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row['Restaurant Name']} | Rating: {row['Aggregate rating']}"
    ).add_to(cluster)

# Save map
file_name = "restaurant_map.html"
m.save(file_name)

# Open automatically in browser
webbrowser.open(file_name)

print("Map opened in browser and saved as restaurant_map.html ")
```
